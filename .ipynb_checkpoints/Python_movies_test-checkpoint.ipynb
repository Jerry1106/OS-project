```python
movies_pop.plot(kind = 'hist', y = 'vote_count', range = (0,4000), bins = 10, normed = False) #bin共幾條條狀，range範圍，normed指密度 true = 1
# 將圖分成1000個長條，vote_count的數值範圍是0~100，出現在x值的頻率是y值 （出現在20左右次數最高）
plt.show()
```

```python
cast_list = list(map(str, movies.cast))
cast = []
for i in cast_list:
    split_cast = list(map(str, i.split('|')))
    for j in split_cast:
        if j not in cast:
            cast.append(j)
```

```python
from collections import Counter
result = Counter(cast)
result
```

```python
movies_vote = movies[['original_title', 'vote_count', 'vote_average']]
movies_vote.head(3)
```

```python
movies_vote.describe()
```

```python
movies_vote = movies_vote.sort_values(by ='vote_average', ascending=False)
movies_vote['vote_status'] = movies_vote.vote_average.map(lambda x: 'High' if x > 6 else 'Low')
movies_vote.corr()
```

```python
plt.figure(figsize = (16,5))
sns.heatmap(movies_vote.corr(), cmap="coolwarm", annot=True)
plt.show()
```
